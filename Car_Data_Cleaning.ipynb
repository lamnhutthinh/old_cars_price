```python
df_unclean_focus['mileage']=df_unclean_focus['mileage'].apply(lambda x: int(str(x).replace(',','')) if str(x).replace(',','').isdigit() else 0)
df_unclean_focus['mileage2']=df_unclean_focus['mileage2'].apply(lambda x: int(str(x).replace(',','')) if str(x).replace(',','').isdigit() else 0)
for i in range(df_unclean_focus.shape[0]):    
    if df_unclean_focus.iloc[i,df_unclean_focus.columns.tolist().index('mileage')] == 0:
        df_unclean_focus.iloc[i,df_unclean_focus.columns.tolist().index('mileage')]=df_unclean_focus.iloc[i,df_unclean_focus.columns.tolist().index('mileage2')]
```

```python
df_unclean_focus=df_unclean_focus[df_unclean_focus['engine size2']!='Unknown']
```

```python
df_unclean_focus['engine size']=df_unclean_focus['engine size'].apply(lambda x: round(float(x.replace('cc','').replace(',','').replace('T','')),1) if str(x).replace('.','').replace(',','').replace('cc','').replace('T','').strip().isdigit() else 0)
df_unclean_focus['engine size2']=df_unclean_focus['engine size2'].apply(lambda x: round(float(x.replace('cc','').replace(',','').replace('T','')),1) if str(x).replace('.','').replace(',','').replace('cc','').replace('T','').strip().isdigit() else 0)
for i in range(df_unclean_focus.shape[0]):    
    if df_unclean_focus.iloc[i,df_unclean_focus.columns.tolist().index('engine size')] == 0:
        df_unclean_focus.iloc[i,df_unclean_focus.columns.tolist().index('engine size')]=df_unclean_focus.iloc[i,df_unclean_focus.columns.tolist().index('engine size2')]
```
